````python
from google.colab import drive
drive.mount('/content/drive')

import os, cv2, numpy as np, glob, gc
import tensorflow as tf

CLASSES   = ["nonviolent", "violent"]
SAVE_PATH = "/content/drive/MyDrive/violence detection ai/numpy_data"
EXPORT_PATH = "/content/drive/MyDrive/violence detection ai/exports"
os.makedirs(EXPORT_PATH, exist_ok=True)

# ── 1. Load best model ──────────────────────────────────────────
model = tf.keras.models.load_model(f"{SAVE_PATH}/best_model.keras")
print("Model loaded!")

# ── 2. Save as .h5 ─────────────────────────────────────────────
model.save(f"{EXPORT_PATH}/violence_model.h5")
print("Saved as .h5!")

# ── 3. Save as SavedModel format (for TFLite/backend) ──────────
model.save(f"{EXPORT_PATH}/violence_model_savedmodel")
print("Saved as SavedModel!")

# ── 4. Save as TFLite (lightweight for mobile/backend) ─────────
converter = tf.lite.TFLiteConverter.from_saved_model(
    f"{EXPORT_PATH}/violence_model_savedmodel"
)
converter.optimizations = [tf.lite.Optimize.DEFAULT]
tflite_model = converter.convert()
with open(f"{EXPORT_PATH}/violence_model.tflite", "wb") as f:
    f.write(tflite_model)
print("Saved as .tflite!")

# ── 5. Inference pipeline ───────────────────────────────────────
SEQUENCE_LENGTH = 16
IMG_SIZE        = 112
THRESHOLD       = 0.9

def extract_frames(video_path):
    cap    = cv2.VideoCapture(video_path)
    total  = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    frames = []
    if total < SEQUENCE_LENGTH:
        cap.release()
        return []
    step = total // SEQUENCE_LENGTH
    for i in range(SEQUENCE_LENGTH):
        cap.set(cv2.CAP_PROP_POS_FRAMES, i * step)
        ok, frame = cap.read()
        if not ok:
            break
        frame = cv2.resize(frame, (IMG_SIZE, IMG_SIZE))
        frame = frame / 255.0
        frames.append(frame)
    cap.release()
    return frames

def predict_video(video_path):
    frames = extract_frames(video_path)
    if len(frames) != SEQUENCE_LENGTH:
        return None, None
    X = np.array([frames], dtype=np.float32)  # (1, 16, 112, 112, 3)
    prob = model.predict(X, verbose=0)[0][0]
    label = "VIOLENT" if prob >= THRESHOLD else "NONVIOLENT"
    return label, float(prob)

print("\nInference pipeline ready!")
print("Usage: label, prob = predict_video('your_video.mp4')")

# ── 6. Test inference pipeline ──────────────────────────────────
test_videos = glob.glob(
    "/content/drive/MyDrive/violence detection ai/dataset/test/violent/*.mp4"
)[:3]

if test_videos:
    print("\nTesting on 3 sample videos:")
    print("-" * 40)
    for v in test_videos:
        label, prob = predict_video(v)
        print(f"{os.path.basename(v)} → {label} ({prob:.4f})")
else:
    print("\nNo test videos found for demo — inference pipeline ready to use!")

# ── 7. Print input contract ─────────────────────────────────────
print("\n" + "=" * 40)
print("       INPUT CONTRACT")
print("=" * 40)
print("Input  : Video file (.mp4, .avi)")
print("Frames : 16 evenly spaced frames")
print("Size   : 112x112 pixels")
print("Normalized : divide by 255.0")
print("Output : Probability 0.0 to 1.0")
print(f"Threshold  : {THRESHOLD}")
print("Classes: 0=Nonviolent, 1=Violent")
print("=" * 40)

# ── 8. Write README to Drive ────────────────────────────────────
readme = """# Violence Detection Model

## Model Info
- Architecture : MobileNetV2 + LSTM
- Input        : 16 frames, 112x112, normalized
- Output       : Probability (0-1), threshold=0.9
- Accuracy     : 89.1%
- F1 Score     : 0.91
- ROC-AUC      : 0.9698

## Files
- violence_model.h5           → TensorFlow model
- violence_model.tflite       → Lightweight model for deployment
- violence_model_savedmodel/  → SavedModel format

## Usage
```python
label, prob = predict_video('video.mp4')
# Returns: ('VIOLENT', 0.95) or ('NONVIOLENT', 0.12)
```

## Input Format
- Video file: .mp4 or .avi
- Min length: 16 frames
- Auto resized to 112x112
- Normalized to 0-1

## Pipeline
Video → Extract 16 frames → Resize 112x112
→ Normalize → MobileNetV2 → LSTM → Probability
"""

with open(f"{EXPORT_PATH}/README.md", "w") as f:
    f.write(readme)
print("README saved!")

print("\n" + "=" * 40)
print("     PHASE 5 COMPLETE!")
print("=" * 40)
print(f"All files saved to: {EXPORT_PATH}")
print("violence_model.h5        ✅")
print("violence_model.tflite    ✅")
print("violence_model_savedmodel✅")
print("README.md                ✅")
print("Inference pipeline       ✅")
print("=" * 40)
````
